## 1. 使用Git访问github代码仓库

> 把Github单独写一个节，是因为github的账号使用了特殊的字符（比如`@`符号,这是邮箱中经常使用的符号），需要使用稍微麻烦一点的登录模式(因为登录账号可以直接嵌入http的链接参数中，但是因为特殊的符号问题，导致这种方式无法使用)。

- 前置条件：

>已经在github系统中创建了一个仓库。
> 

- github的http访问协议地址;
    - `https://github.com/QiangAI/AI_Engineering.git`

### 1.1. 克隆github代码仓库到本地

- 命令：`git clone https://github.com/QiangAI/AI_Engineering.git`

- 执行截图
    - 

- 克隆下来的仓库就是一个本地仓库，可以直接在仓库的工程中工作，工作后的结果，可以提交同步到github服务器。

### 1.2. push代码到github服务器

- 把本地仓库的工作同步到github服务器需要账号密码，登录有两种方式：
    1. 免密登录，这个需要使用ssh工具生成私钥与公钥，并把公钥拷贝配置到github服务器中，这个过程配置麻烦，后面节省时间，提升效率。但是第一次配置麻烦。
    2. 第一次push代码的时候，提供账号与口令，后面就不再提供账号与口令；我喜欢这种方式。
    3. 提示：其实还有一种方式，使用http的账号与口令机制，在http的url中设置账号与口令。

- 操作摘要

> - 条件：在克隆的本地仓库下工作：
    >1. 设置用户名与用户邮箱
    >2. 添加改变的文件到本地仓库
    >3. 添加注释到本地仓库
    >4. 上传本地仓库文件到远程服务器（会提示输入口令：Token）

1. 设置用户名与用户邮箱

- 设置的用户名，一定是github指定的用户名与邮箱。
    - 

- 操作命令：
    - `git config --global user.name "QiangAI"`
    - `git config --global user.email "38395870@qq.com"`

- 查看本地的user设置
    - `git config --global user.name`
    - `git config --global user.email`
    - 

- 查看所有配置
    - `git config --list`

2. 添加改变的文件到本地仓库

- 在本地仓库上工作后,可以添加本地文件到本地仓库
    - 命令：`git add  *`
    - `*`表示当前目录下所有文件与目录。也可以使用具体的文件与目录名，或者其他文件通配符：`?`等。

3. 添加注释到本地仓库

4. 上传本地仓库文件到远程服务器
